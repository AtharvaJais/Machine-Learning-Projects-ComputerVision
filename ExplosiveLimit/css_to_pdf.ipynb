```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.lib.colors import Color
from PIL import Image

def exportpdf():
    class PDFExporter:
        
        def __init__(self, file_path):
            self.file_path = file_path
            self.canvas = canvas.Canvas(file_path, pagesize=letter)

        def add_element(self, x, y, width, height, image_path):
            self.canvas.drawInlineImage(image_path, x, y, width, height)

        def save(self):
            self.canvas.save()

    def main():
        pdf_exporter = PDFExporter("sample_report.pdf")
        elements = [(-20, 460, 655, 195, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/test_info_chart.png"),
                    (220, 640, 165, 125, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/summary_chart.png"),
                    (46, 385, 120, 110, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/avg_p_chart.png"),
                    (170, 375, 150, 120, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/diff_lvl_chart.png"),
                    (305, 375, 140, 120, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/std_grade_chart.png"),
                    (450, 377, 150, 120, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/bt_std_dist_chart.png"),
                    (17, 100, 320, 285, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/top_ten_std_bt_yr_chart.png"),
                    (300, 100, 320, 285, "/home/atharva/Documents/ComputerVision/ExplosiveLimit/pdf_images/bottom_ten_std_bt_yr_chart.png")
                   ]
        for element in elements:
            pdf_exporter.add_element(*element)
        pdf_exporter.save()
    if __name__ == "__main__":
        main()

exportpdf()
```
