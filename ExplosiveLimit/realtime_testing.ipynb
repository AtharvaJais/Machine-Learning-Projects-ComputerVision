```python
import cv2
import numpy as np
from keras.models import load_model


model_final = load_model("/home/atharva/Documents/ComputerVision/ExplosiveLimit/trained_model/plastic_container_model.h5")
camera = cv2.VideoCapture(0)

while True:
    ret, frame = camera.read()
    resized_frame = cv2.resize(frame, (224, 224), interpolation=cv2.INTER_AREA)
    img = np.expand_dims(resized_frame, axis=0)
    prediction = model_final.predict(img)
    if prediction[0][0] > prediction[0][1]:
        cv2.putText(frame, "Positive: Plastic container present", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    else:
        cv2.putText(frame, "Negative: No plastic container", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

    cv2.imshow("Real-time Object Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
